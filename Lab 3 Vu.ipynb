```python
file1 = open("romeoandjuliet.txt", "r")
```

```python
contents = file1.read().lower()
print(contents)
```

```python
# remove all punctuations 

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

fcontents = ""
for char in contents:
    if char not in punctuations:
        fcontents += char
print(fcontents)
```

```python
# remove all stop words

import io
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

stop_words = set(stopwords.words('english'))
```

```python
print(stop_words)
print(len(stop_words))
```

```python
#split the text from fcontents into words

dwords = fcontents.split()
print(dwords)
```

```python
for r in dwords: 
    if not r in stop_words: 
        appendFile = open('filteredromeoandjuliet.txt','a') 
        appendFile.write(" "+r) 
        appendFile.close()
```

```python
file2 = open('filteredromeoandjuliet.txt', 'r')
contents2 = file2.read()
print(contents2)
```
